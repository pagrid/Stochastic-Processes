```python
# -*- coding: utf-8 -*-
"""
2D_Brownian_Motion.ipynb

Brownian Motion Simulation (2D)
--------------------------------
Author: Petros Agridos
Description:
Simulates the 2D Brownian motion of many particles, visualizes their
random trajectories, and verifies Einstein’s diffusion law:
⟨r²(t)⟩ = 4 D t
"""

import numpy as np
import matplotlib.pyplot as plt

# Parameters
N_particles = 5000
N_steps = 1000
dt = 1e-2
D = 0.5
sigma = np.sqrt(2 * D * dt)

# Initialize positions
x = np.zeros((N_particles, N_steps))
y = np.zeros((N_particles, N_steps))

# Random walk simulation
for i in range(1, N_steps):
    dx = np.random.normal(0, sigma, N_particles)
    dy = np.random.normal(0, sigma, N_particles)
    x[:, i] = x[:, i - 1] + dx
    y[:, i] = y[:, i - 1] + dy

# Compute mean square displacement <r^2>
r2 = np.mean(x**2 + y**2, axis=0)
time = np.arange(N_steps) * dt

# ---- Plot 1: Sample trajectories ----
plt.figure(figsize=(7, 7))
for i in range(400):
    plt.plot(x[i, :], y[i, :], lw=0.8)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Brownian Motion: Sample Trajectories")
plt.axis("equal")
plt.grid(alpha=0.3)
plt.show()

# ---- Plot 2: Mean square displacement ----
plt.figure(figsize=(8, 6))
plt.plot(time, r2, 'o', label=r"Simulation $\langle r^2 \rangle$")
plt.plot(time, 4 * D * time, 'r--', label=r"Theory $4Dt$")
plt.xlabel("Time t")
plt.ylabel(r"Mean square displacement $\langle r^2(t) \rangle$")
plt.title("Einstein’s Diffusion Law Verification (2D)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
